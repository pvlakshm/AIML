```python
from sklearn.datasets import make_blobs

# 1. Generate synthetic dataset
X, y_true = make_blobs(
    n_samples=300, 
    centers=3, 
    cluster_std=1.0, 
    random_state=42
)
```

```python
from sklearn.cluster import KMeans

# 2. Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)
```

```python
import matplotlib.pyplot as plt

# 3. Plot original dataset (true blobs)
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap="viridis", s=40, edgecolor="k")
plt.title("Original dataset (true blobs)")

# 4. Plot KMeans result
plt.subplot(1,2,2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap="viridis", s=40, edgecolor="k")
plt.scatter(kmeans.cluster_centers_[:, 0], 
            kmeans.cluster_centers_[:, 1], 
            s=200, c="red", marker="X", label="Centroids")
plt.title("K-means clustering result")
plt.legend()

plt.show()
```
