```python
# We have price data for 20 houses.
# The houses have 3 features:
#   size: 1400 to 3500 sqft
#   bedrooms: mostly 4â€“5, some 3s and 6s
#   age: decreasing from 20 years to 1 year
#
# The prices increase from $250K to $633K, as follows:
#   as size increases, the price increases
#   as num. of bedrooms increases, the price increases
#   as age increase, the price decreases
#
# In this case, we have calculated the price (in $1000s) using this formula:
#   price = 50 + 0.15*size + 10*bedrooms - 2*age
#
# The goal of this exercise is to learn this formula!

# Step 1: create the data points
import numpy as np
# Features: [size (sqft), bedrooms, age (years)]
X = np.array([
    [1400, 3, 20],
    [1600, 3, 15],
    [1700, 4, 10],
    [1800, 4,  8],
    [2000, 4, 12],
    [2100, 5,  5],
    [2200, 4,  9],
    [2300, 4,  4],
    [2400, 5,  5],
    [2500, 5,  7],
    [2600, 4,  3],
    [2700, 5,  2],
    [2800, 5,  6],
    [2900, 5,  1],
    [3000, 4,  4],
    [3100, 4,  3],
    [3200, 5,  2],
    [3300, 5,  1],
    [3400, 5,  2],
    [3500, 6,  1]
])

# Prices
y = np.array([
    250, 290, 325, 344, 366, 405, 402, 427, 450, 461,
    474, 501, 508, 533, 532, 549, 576, 593, 606, 633
])
```

```python
# Step 2: Visualize the data
import matplotlib.pyplot as plt
plt.plot(y, color = 'blue', marker = 'v', label = 'Actual Price')
plt.xlabel('House Index')
plt.ylabel('Price ($1000s)')
plt.title('Actual House Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 3: Fit a Multivariate Linear Regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)

# print the learned intercept and coefficients.
# it matches with "50 + 0.15*size + 10*bedrooms - 2*age". How cool is that!
print("Intercept:", np.round(model.intercept_, 2))
print("Coefficients:", np.round(model.coef_, 2))
```

```python
# Step 4: Make predictions
# Use the same X (features). Since the model learned the formula, the predictions
# will exactly match.
y_pred = model.predict(X)
```

```python
# Step 5: Visualize
plt.plot(y, color = 'blue', marker = 'v', label = 'Actual Price')
plt.plot(y_pred, color = 'red', marker = '^', label = 'Predicted Price')
plt.xlabel('House Index')
plt.ylabel('Price ($1000s)')
plt.title('Multivariate Linear Regression: House Price Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
